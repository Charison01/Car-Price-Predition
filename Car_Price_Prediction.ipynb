### Additional Data Cleaning Steps

```python
# Getting information on the column data types
car_df.info()
```

```python
# Function that converts strips and converts non-numerical data types to numericals.
def replace_data(data, to_replace, value, col):
    """Replaces values in the given column and converts to numerical"""
    data[col].replace(to_replace, value,inplace=True, regex = True)
    data[col] = pd.to_numeric(data[col], errors='coerce')
    return data
```

```python
# Engine column: calling function for stripping and converting to numerical
replace_data(data=car_df, to_replace="CC",value="",col="engine")
```

```python
# Mileage column: calling function for stripping and converting to numerical
replace_data(data=car_df, to_replace="kmpl" or "km/kg",value="",col="mileage")
```

```python
# Max_power column: calling function for stripping and converting to numerical
replace_data(data=car_df, to_replace="bhp",value="",col="max_power")
```

We use the `re` library to perform regular expressions on the column torque. A function is employed to perform the fuctionality and extract relevant information from the torque column.

```python
# Functiong for Regular Expression: Extracting numerical data from a string by finding patterns 
def find_first_float(string):
    value_obtained = re.findall(r"[-+]?(?:\d*\.*\d+)",string)[0]
    return value_obtained
```

```python
import os
import contextlib
with open(os.devnull, "w") as f, contextlib.redirect_stdout(f):
    car_df["torque"] = car_df["torque"].apply(find_first_float).astype(float);
```
